```python
# Importing Libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set(color_codes=True)
```

```python
# Reading csv
df = pd.read_csv("titanic.csv")
df.head()
```

```python
# Checking shape rows and columns
print(df.shape, " :Rows and columns")
```

```python
# Data info
df.info()
```

```python
# Visualizing empty data
"""
Cabin is mostly having null values so it is better than we remove that feature completly while we need to handle Age feature
differently.
"""

sns.heatmap(df.isnull(), yticklabels=False, cbar=False, cmap='tab20_r')
plt.title('Missing Data')
plt.show()
```

```python
# Checking Age by Passenger Class
plt.figure(figsize=(10, 7))
sns.boxplot(x="Pclass", y="Age", data=df).set_title("Age by Passenger Class")
plt.show()
```

```python
'''
Now we are replacing the null values in age into there median we found at the above diagram.
'''

def impute_age(cols):
    Age = cols[0]
    Pclass = cols[1]
    
    if pd.isnull(Age):
        
        if Pclass == 1:
            
            return 37
        
        elif Pclass == 2:
            
            return 29
        
        else:
            
            return 24
    else:
       return Age
        
'''
Applying the function to the age column, on basis of Pclass if Age value is null or NaN it would be replaced by 
the following number
'''

df['Age'] = df[['Age', 'Pclass']].apply(impute_age, axis=1)
```

```python
# Remove Cabin Feature
df.drop('Cabin', axis = 1, inplace = True)
df.dropna(axis=0, how="any", inplace=True)
```

```python
# Visualizing empty data
"""
Checking if null value still exists or not.
"""

print(df.shape, " :Rows and columns")
sns.heatmap(df.isnull(), yticklabels=False, cbar=False, cmap='tab20_r')
plt.title('Missing Data')
plt.show()
```

```python
'''
Droping Name, Ticket Features because they are string values and it will not help in model training even if
we convert them into numbers so we will drop them.
'''

df.drop(columns=['Name', 'Ticket'], inplace=True)
```

```python
'''
Sex and Embarked are categories here, so we convert objects to category datatype :) As it will help further when we will
do encoding, we will just need to write a small function than.
'''

objToCategory = ['Sex', 'Embarked']

for colname in objToCategory:
    df[colname] = df[colname].astype('category')
```

```python
df.describe()
```

```python
'''
Convert categorical features i.e Sex and Embarked into dummy variables or indicator variables
'''

sex = pd.get_dummies(df['Sex'], drop_first = True)
embarked = pd.get_dummies(df['Embarked'], drop_first = True)
```

```python
# Orignal Data after dropping NA features and Obervations(rows)
df.head()
```

```python
'''
Concating the indicator variables in the dataframe features Sex and Embarked, we can see the male, Q and S in next step 
we will drop the original Sex and Embarked Column
'''

df = pd.concat((df, sex, embarked), axis = 1)
df.head()
```

```python
# Now we can see that all the data is in numbers now we can train the model. Data processing is done yaaaah!!!
df.drop(['Sex', 'Embarked'], axis=1)
```

```python
# Dividing data into X(independent variable) axis and Y(Dependent variable/target variable) axis
# Also axis =1 says that drop the given column/feature

X = df.drop(["Survived"], axis = 1)
y = df["Survived"]
```

```python
# splitting data into test and train data
'''
Random state is getting random data from the dataset this is because if data is not random than let's say in the start of
data most of people are alive than model will not be able to predict whether the person is dead or not this will decrease model
accuracy, as model will only be trained to predict one class of data better. If random_state is not set than for example from a 
training data 60% people will be alive while last 10% will be dead here model will be trained more on a specific class
meaning model accuracy will be biased. By random_state we will get random data, the more the value is the more the data will be
random.
'''

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state = 101)

# Fun fact X is always input while Y is output/target-variables this is a convention followed by whole ML community.
```

```python
# Fit model, we are using LogisticRegression for finding the survived gender
model = LogisticRegression()
model.fit(X_train, y_train)
```
